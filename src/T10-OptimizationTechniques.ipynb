# Momentum Optimization

- What type of problem does momentum solve?
  - It is Non-Convex optimizer.
  - So, it solve high curvature, consistent gradient(where slope changes very slow) and Noisy Gradient(bcz of which we face local minima)
- Intuition:
  - If the previous gradient sending u in one direction then confidently increasing the speed move in that direction(that is why, the single most important use of momentum optimization is speed) else if the previous some gradient sending u in forwar and some in backward the we decrease the speed and move accordingly.</br>
   Here beta is decaying factor</br>
   For more [info](https://www.youtube.com/watch?v=vVS4csXRlcQ&list=PLKnIA16_RmvYhD5pqAeVu3j_jTjnTJIW2&index=3&ab_channel=CampusX)</br>
   
   
- the disadvantage of Momentum here is when it go to global minima due to momentum it start oscillating and then settle down, this oscillations takes time, although it is greater than SDG but with upcomming optimizers its slow. So it won't get the fastest optimization technique. And this is the reason it new Optimizers came.

- so here the bigest problem of Momentum Optimizations is momentum itself.
